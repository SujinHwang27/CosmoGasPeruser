# Discrete Cosine Transform 
This notebook explores usages of Discrete Cosine Transform (DCT). Suppose we have a binary class dataset showing same trends but with different noises (detail). 

```python
import numpy as np
from scipy.fftpack import dct
import matplotlib.pyplot as plt

# Two similar signals with small differences (noise)
data1 = np.array([10, 20, 30, 40, 50, 45, 35, 25], dtype=float)
data2 = np.array([11, 19, 31, 39, 51, 44, 36, 26], dtype=float)

# Compute DCT (type-II, orthonormal)
dct1 = dct(data1, type=2, norm='ortho')
dct2 = dct(data2, type=2, norm='ortho')

# Plot original signals
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.plot(data1, 'o-', label='Data 1')
plt.plot(data2, 's--', label='Data 2')
plt.title("Original Signals")
plt.legend()

# Plot DCT coefficients
plt.subplot(1,2,2)
plt.plot(dct1, 'o-', label='DCT Data 1')
plt.plot(dct2, 's--', label='DCT Data 2')
plt.title("DCT Coefficients")
plt.xlabel("Coefficient Index (Frequency)")
plt.ylabel("Magnitude")
plt.legend()
plt.tight_layout()
plt.show()

# Print DC and high-frequency comparison
print(f"DC term comparison: {dct1[0]:.2f} vs {dct2[0]:.2f}")
print(f"High-frequency term comparison (last coefficient): {dct1[-1]:.2f} vs {dct2[-1]:.2f}")

```
