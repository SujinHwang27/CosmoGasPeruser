```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npy files
flux_array = np.load("all_fluxes.npy")  # Assuming this is the file for flux data
SN_array = np.load("all_SN_values.npy")  # Assuming this is the file for SN values

# Check the shapes for consistency
if flux_array.shape[0] != len(SN_array):
    raise ValueError("Mismatch between the number of flux curves and SN values.")

# Split the indices into two groups
midpoint = len(SN_array) // 2
group_1_indices = range(0, midpoint)  # First half
group_2_indices = range(midpoint, len(SN_array))  # Second half

# Plot group 1
plt.figure(figsize=(10, 6))
for i in group_1_indices:
    plt.plot(flux_array[i], label=f"SN = {SN_array[i]}")
plt.ylabel("Flux", fontsize=12)
plt.title("Flux Curves for Lower Signal-to-Noise Ratios", fontsize=14)
plt.legend(title="Signal-to-Noise Ratios", fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()

# Plot group 2
plt.figure(figsize=(10, 6))
for i in group_2_indices:
    plt.plot(flux_array[i], label=f"SN = {SN_array[i]}")
plt.ylabel("Flux", fontsize=12)
plt.title("Flux Curves for Higher Signal-to-Noise Ratios", fontsize=14)
plt.legend(title="Signal-to-Noise Ratios", fontsize=10)
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Load the .npy files
flux_array = np.load("all_fluxes.npy")  # Assuming this is the file for flux data
SN_array = np.load("all_SN_values.npy")  # Assuming this is the file for SN values

# Check the shapes for consistency
if flux_array.shape[0] != len(SN_array):
    raise ValueError("Mismatch between the number of flux curves and SN values.")

# Loop through each flux curve and plot it individually
for i, sn in enumerate(SN_array):
    plt.figure(figsize=(10, 4))
    plt.plot(flux_array[i], label=f"SN = {sn}")
    plt.ylabel("Flux", fontsize=12)
    plt.title(f"Flux Curve for Signal-to-Noise Ratio SN = {sn}", fontsize=14)
    plt.legend(fontsize=10)
    plt.grid(alpha=0.3)
    plt.tight_layout()
    plt.show()
```
